# finding ansqers with the data we have

```python
# the distribution of ratings
sns.countplot(x='rating',data=df,color='c')
```

**What was the best year of sales**

```python
df['year']=pd.DatetimeIndex(df['timestamp']).year
```

```python
df.groupby('year')['rating'].count().plot(kind='bar')
```

```python
#What brand sold the most in 2015
df_2015=df[df['year']==2015]
```

```python
df_2015.groupby('brand')['rating'].count().sort_values(ascending=False).head(10).plot(kind="bar")
```

### we can see that the brand name of Bose sold the most followed closely with mpow while the least sold was fintie

```python
#how much was made sales in the year 2015
df[df['year']==2015].groupby('year')['rating'].count().plot(kind="bar")
```

## we can see that the year 2025 had the best  sales.

```python
# what was the best month of sales
df['month']=pd.DatetimeIndex(df['timestamp']).month
```

```python
df.groupby('month')['rating'].count().plot(kind="bar")
```

### the month of january had the best sales.

```python
# what product sold the most in 2016
df[df['year']==2016].groupby('brand')['rating'].count().sort_values(ascending=False).head(10).plot(kind="bar",color='c')
```

**The top 3 products sold in 2016 were bose ,logitech,taotronics**

```python
# which product sold in 2017
df[df['year']==2017].groupby('brand')['rating'].count().sort_values(ascending=False).head(10).plot(kind="bar",color='pink')
```

**the top 3 products sold in 2017 were bose,logitech,mpow**

```python
# which product by category sold the most
df.groupby('category')['rating'].count().sort_values(ascending=False).head(10).plot(kind='barh')
```

### we can see that the category of Headphones  sold the most
### cumputers and accessories were sold the second most
### camera & photo sold the third most followed by accesories and supplies
### the least sold category was security and surveillance

```python
#which product by brand name sold the least
df.groupby('brand')['rating'].count().sort_values().head(10).plot(kind='bar',color='magenta')
```
